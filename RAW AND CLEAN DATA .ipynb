```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

## HYDERABAD :

```python
names = []
address = []
city = []
features = []
ratings = []
votes = []
prices = []
for i in range(1,6):
        url="https://www.oyorooms.com/hotels-in-hyderabad/?page="+str (i)
        print(url)
        
        page = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
        soup = BeautifulSoup(page.text, 'html.parser')
        
        N = soup.find_all("a", class_="c-nn640c u-width100")
        for i in N:
            a = i.text
            names.append(a)
        
        rat=soup.find_all('div',class_="oyo-cell--12-col listingHotelDescription__content")
        for i in rat:
            a = i.text
            b = re.findall("(\d\.\d)\s\(", a)
            ratings.append(b) 
        
        pr=soup.find_all("span",class_='listingPrice__finalPrice')
        for i in pr:
            a=i.text
            prices.append(a)  
       
        add=soup.find_all("span",class_="u-line--clamp-2")
        for i in add:
            a=i.text
            address.append(a.replace(', Hyderabad', ""))  
            
            
        vo=soup.find_all("div",class_="oyo-cell--12-col listingHotelDescription__content")
        for i in vo:
            a =i.text
            b = re.findall("\((\d+)\sRatings", a)
            votes.append(b)
        ci=soup.find_all("span",class_="u-line--clamp-2")
        for i in ci:
            a=i.text
            city.append(re.findall(r"Hyderabad", a)) 
            
            
        fea = soup.find_all("div", class_="amenityWrapper")
        for i in fea:
            a = i.text
            features.append(a.split("+")[0])
```

```python
print(len(names))
print(len(ratings))
print(len(prices))
print(len(address))
print(len(votes))
print(len(city))
print(len(features))
```

```python
dic={"Hotel_Names":names,
   "Address":address,
   "City":city,
   "Ratings":ratings,
   "Votes":votes,
   "Prices":prices,
   "Features":features
  }
```
